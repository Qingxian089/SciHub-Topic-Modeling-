```python
import os

# get the file path
def list_dir(file_dir):
    list_csv = []
    dir_list = os.listdir(file_dir)
    for cur_file in dir_list:
        path = os.path.join(file_dir,cur_file)
        if os.path.isfile(path):
            dir_files = os.path.join(file_dir, cur_file)
        if os.path.splitext(path)[1] == '.csv':
            csv_file = os.path.join(file_dir, cur_file)
            list_csv.append(csv_file)
        if os.path.isdir(path):
            list_dir(path)
    return list_csv

list_csv1 = list_dir('/dbfs/mnt/lsde/group14/data_1-1000')
list_csv2 = list_dir('/dbfs/mnt/lsde/group14/data_1001-1500')
list_csv3 = list_dir('/dbfs/mnt/lsde/group14/data_1501-2500')
list_csv4 = list_dir('/dbfs/mnt/lsde/group14/data_2501-2999')
list_csv5 = list_dir('/dbfs/mnt/lsde/group14/data_3000-4001')
list_csv6 = list_dir('/dbfs/mnt/lsde/group14/data_4002-4746')
# get all the paths
list_csv = list_csv1+list_csv2+list_csv3+list_csv4+list_csv5+list_csv6
```

```python
import pandas as pd
# read data divied by areas
clean = pd.read_csv('/dbfs/mnt/lsde/group14/Data_with_Area1.csv', names=['year', 'title', 'subject', 'content', 'area'])
print(clean)
```

```python
# read data by area
clean1 = clean.loc[(clean['area'] == 'Math')]
clean2 = clean.loc[(clean['area'] == 'Medicine')]
clean3 = clean.loc[(clean['area'] == 'Chemistry')]
clean4 = clean.loc[(clean['area'] == 'Biology')]
clean5 = clean.loc[(clean['area'] == 'Humanites')]
clean6 = clean.loc[(clean['area'] == 'Physics')]
clean7 = clean.loc[(clean['area'] == 'Engineering')]
clean8 = clean.loc[(clean['area'] == 'Ecology')]
clean9 = clean.loc[(clean['area'] == 'Computer Science')]
clean10 = clean.loc[(clean['area'] == 'GeoSci')]

frames = [clean1, clean2, clean3, clean4, clean5, clean6, clean7, clean8, clean9, clean10]

# integrating Data
final = pd.concat(frames,ignore_index=True)
final.to_csv('/dbfs/mnt/lsde/group14/area_final.csv')
```

```python
# group papers by area
groups = clean.groupby("area")
# Determine the size of each group
counts = groups.size()
# count the papers by area and reset the index
df = pd.DataFrame({'count' : clean.groupby( [ "area" ] ).size()}).reset_index()
df.to_csv('/dbfs/mnt/lsde/group14/area_count.csv')
```
