```python
import re

# get the ASJC of a journal
def Query(title):
    title = title.lower()
    return Journal2ASJC[Journal2ASJC['SourceTitle'] == title]   
#Journal2ASJC.loc[3,'All Science Journal Classification Codes (ASJC)'][6:10]

#get the area from the ASJC
def Area(code2):
    return sub[sub['ASJC'] == int(code2)]

#get the asjc code for jounal, then wirte the discipline and the area
def Classify(data):
    for k in range(0,len(data)):
        try:
            text = Query(data['subject'][k])
            asjc = text['All Science Journal Classification Codes (ASJC)'].iloc[0]
#             print(asjc)
            sbnm = int((len(asjc)+1)/6)
#             print(sbnm)
            for i in range(0,sbnm):
                code = int(asjc.split(";",k)[i])
#                 area = Area(asjc)['Area']
                area = Area(code)['Area']
                area = str(area).split("\n")[0].split("    ")[1]
#                 print(area)
                sd = Area(code)['Sub-Discipline']
                sd = str(sd).split("\nName: ")[0].split("    ")[1]
#                 print(sd)
                if data.loc[k,'area'] == "":
                    data.loc[k,'area'] = area
                    data.loc[k,'subdiscipline'] = sd + ";"
                elif str(area) == data.loc[k,'area']:
                    if str(sd) != str(data.loc[k,'subdiscipline']).split(";")[0]:
                        data.loc[k,'subdiscipline'] = data.loc[k,'subdiscipline'] + sd + ";"
                    else:
                        print('repeat')
                else:
                    print('sth wrong')
                    pass
                print('suc')
        except:
            print('error!')

print(data.loc[102,'area'])   
print(data.loc[102,'subdiscipline'])
```
