```python
# Import required libraries
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from io import BytesIO
import zipfile
import io
import tika
from tika import parser 
import requests
import json

# Create SparkSession for Spark DataFrame operation
spark = SparkSession.builder.appName('ReadCompressedPDF').getOrCreate()
# Get SparkContext to access RDDs
sc = spark.sparkContext
```

```python
def zip_extract(x):
      """
      Extract PDF contents from a zip file

      Args:
        x: List of [zip file path, zip file content]

      Returns:
        List of tuples (pdf path, pdf content) 
      """
    in_memory_data = BytesIO(x[1])
    file_obj = zipfile.ZipFile(in_memory_data, "r")
    names = [i for i in file_obj.namelist()]
    files = names[1:]  
    return zip(files, [file_obj.open(file).read() for file in files])

def extract_content(pdf):
      """
      Extract metadata and text content from a PDF

      Args:
        pdf: Tuple of (PDF path, PDF content)

      Returns:
        title, doi, create time, text content
      """
    content = pdf[1]
    parsed = parser.from_buffer(content)
    metadata = parsed["metadata"]
    # Extract DOI
    doi = metadata.get("WPS-ARTICLEDOI")
    if not doi:
        print("ARTICLE DOI missing")
        doi = metadata.get("WPS-JOURNALDOI")
        if not doi:
            print("JOURNAL DOI missing")
    title = metadata.get("dc:title")
    createtime = metadata.get("dcterms:created")  
    return title, doi, createtime, parsed["content"]
```

```python
# Load ZIP files as binary RDD
zips_rdd = sc.binaryFiles("dbfs:/FileStore/tables/*.zip")
# Extract PDF files from zip files and parse PDF contents
pdfs = zips_rdd.flatMap(zip_extract).map(extract_content)


# Convert extarxtion result to DataFrame
columns = ["title","doi", "createtime", "content"]
df = pdfs.toDF(columns)
```

```python
# Define UDF
@udf(returnType=ArrayType(StringType()))
def fetchsubj(doi, **kwargs):
    """Fetch subject for a given DOI.
    Parameters
    ----------
    doi : str
    **kwargs :
        Additional keyword arguments are passed to `json.loads()` if `fmt`
        is 'dict' and you're a big JSON nerd.
    Returns
    -------
    out : stror None
        `None` is returned if the server gives unhappy response. Usually 
        this means the DOI is bad.
    Examples
    --------
    >>> fetchmeta('10.1016/j.dendro.2018.02.005')
    
    References
    ----------
    https://gist.github.com/brews/8d3b3ede15d120a86a6bd6fc43859c5e#file-fetchmeta-py
    """

    # Request data from server.
    url = "https://dx.doi.org/" + str(doi)
    accept_type = 'application/citeproc+json'
    header = {'accept': accept_type}
    r = requests.get(url, headers=header)

    # obtain subject if server response is good.
    out = None
    if r.status_code == 200:
        metadata = json.loads(r.text, **kwargs)
        subject = metadata["subject"]
        if subject:
            out = subject
        else:
            print("----Accessing subject failed----")
    else:
        print('----Accessing metadata failed----')

    return out
```

```python
# Fetch subject for each scientific paper and store in new column "doi"
df = df.withColumn("subject", fetchsubj("doi"))
df.printSchema()
df.show()
```
